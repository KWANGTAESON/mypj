```python
import pandas as pd
all_all = pd.read_csv('전국장애현황수.csv')
all_eq = pd.read_csv('장비.csv')
all_warn = pd.read_csv('경보발생횟수.csv')
```

```python
import matplotlib.font_manager as fm

font_list = [font.name for font in fm.fontManager.ttflist]
```

```python
for i in font_list:
    print(i)
```

```python
all_all = all_all[['지역', 'count']]
```

```python
all_all.to_csv('전국장애현황수.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# 시각화 한글폰트 설정
# plt.rc('font', family='Malgun Gothic')
# sns.set(font="Malgun Gothic",#"NanumGothicCoding", 
# rc={"axes.unicode_minus":False}, # 마이너스 부호 깨짐 현상 해결
# style='darkgrid')
plt.rcParams['font.family'] = 'Apple SD Gothic Neo'
sns.barplot(all_all, y=all_all['지역'], x=all_all['count'] )
plt.title('전국 경보 발생 현황')

plt.show
```

```python
all_eq = all_eq[['Unnamed: 0',	'Unnamed: 1']]
```

```python
all_eq.rename(columns={'Unnamed: 0':'장비명', 'Unnamed: 1':'발생횟수'}, inplace=True)
```

```python
all_eq.to_csv('장비.csv', index=False)
```

```python
plt.pie(all_eq['발생횟수'], labels=all_eq['장비명'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette('tab10'))
plt.title('장비별 경보 비율')
plt.show()
```
